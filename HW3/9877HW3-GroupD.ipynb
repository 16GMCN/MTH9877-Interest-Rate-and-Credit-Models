## Problem 1

The value of a Receiver Swap at time 0 is 

$$V_{swap} = K \sum_{1\leq j \leq n_c} \alpha_j \, P(0,T_{val},T_j^c) - \sum_{1\leq j \leq n_f} \delta_j L_j P(0,T_{val},T_j^f)$$

The value of a receiver swaption is given by the right to enter a swap at the expiration date $T_0$.
$$
\begin{eqnarray}
V_{swaption} &=& \left(K \sum_{1\leq j \leq n_c} \alpha_j \, P(T_0,T_0, T_j^c) - \sum_{1\leq j \leq n_f} \delta_j L_j P(T_0,T_0, T_j^f)\right)^+\\
&=& \left(K\, A(T_0,T_0,T_0,T) - \sum_{1\leq j \leq n_f} \delta_j L_j P(T_0,T_0, T_j^f)\right)^+\\
\end{eqnarray}
$$

The break-even forward swap rate is as follows,

$$S(T_0, T_0, T) = \frac{\sum_{1\leq j \leq n_f} \delta_j L_j P(T_0,T_0, T_j^c)}{A(T_0,T_0,T_0,T)}$$

Thus the value of the receiver swaption at time $T_0$ with the units of annuity is

$$V(T_0) = \frac{V_{swaption}[T_0]}{A(T_0,T_0,T_0,T)} = \left( K-S(T_0, T_0, T) \right)^+$$

Take expectation in the swap martingale measure, we can get:

$$V(0) = \frac{V_{swaption}[0]}{A_0(T_0,T)} = E^{Q_{T_0,T}}\left[\frac{V_{swaption}[T_0]}{A(T_0,T_0,T_0,T)}\right] = E^{Q_{T_0,T}}[\left( K-S(T_0, T) \right)^+]$$

Thus we have proved the expression.

### Normal model

From leture note#5 equation 49 and 50 we get the implied normal volatility and other expressions as:
$$
\begin{align*}
    \sigma_n(T, K, F_0, \sigma_0, \alpha, \beta, \rho) &= \alpha\frac{F_0-K}{D(\zeta)} \{ 1 + G(F_{mid})\epsilon + \cdots\} \\
    F_{mid} &= \frac{F_0 + K}{2} \\
    C(F) &= F^\beta \\
    \gamma_1 &= \frac{C'(F_{mid})}{C(F_{mid})} = \frac{\beta}{F_{mid}}\\
    \gamma_2 &= \frac{C"(F_{mid})}{C(F_{mid})} = \frac{\beta(\beta-1)}{F^2_{mid}} \\
    G(F_{mid}) &= \frac{2\gamma_2-\gamma_1^2}{24}[\frac{\sigma_0C(F_{mid})}{\alpha}]^2 + \frac{\rho\gamma_1}{4}\frac{\sigma_0C(F_{mid}）}{\alpha} + \frac{2-3\rho^2}{24} \\
    D(\zeta) &= log(\frac{\sqrt{1-2\rho\zeta+\zeta^2}+\zeta-\rho}{1-\rho})\\
    \zeta &= \frac{\alpha}{\sigma_0(1-\beta)}(F_0^{1-\beta} - K^{1-\beta})
\end{align*}
$$

When ATM, the $\sigma_n$ goes to $\frac00$, and thus we can apply l’Hopital’s rule.

$$
\begin{align*}
    \sigma_n(T, K, F_0, \sigma_0, \alpha, \beta, \rho)_{ATM} &= \alpha\{ 1 + G(F_{mid})\epsilon + \cdots\} \frac{F_0-K}{D(\zeta)} \\
    &= \alpha\{ 1 + G(F_{mid})\epsilon + \cdots\} \frac{\frac{d(F_0-K)}{dK}}{\frac{dD(\zeta)}{dK}}
\end{align*}
$$

$$
\begin{align*}
    \frac{dD(\zeta)}{dK} &= \frac{dD(\zeta)}{d\zeta}\frac{d\zeta}{dK} = -\frac{1}{\sqrt{1-2\rho\zeta+\zeta^2}} \frac{\alpha}{\sigma_0}K^{-\beta}\\
    \frac{dD(\zeta)}{d\zeta} &= \frac{1-\rho}{\sqrt{1-2\rho\zeta+\zeta^2}+\zeta-\rho}\frac{\frac{-\rho+\zeta}{\sqrt{1-2\rho\zeta+\zeta^2}}+1}{1-\rho} = \frac{1-\rho}{\sqrt{1-2\rho\zeta+\zeta^2}+\zeta-\rho}\frac{\frac{-\rho+\zeta + \sqrt{1-2\rho\zeta+\zeta^2}}{\sqrt{1-2\rho\zeta+\zeta^2}}}{1-\rho} = \frac{1}{\sqrt{1-2\rho\zeta+\zeta^2}} \\
    \frac{d\zeta}{dK} &= \frac{\alpha}{\sigma_0(1-\beta)}[- (1-\beta)K^{-\beta}] = -\frac{\alpha}{\sigma_0}K^{-\beta}
\end{align*}
$$

$$
\begin{align*}
    \sigma_n(T, K, F_0, \sigma_0, \alpha, \beta, \rho)_{ATM} &= \alpha\{ 1 + G(F_{mid})\epsilon + \cdots\} \frac{\frac{d(F_0-K)}{dK}}{\frac{dD(\zeta)}{dK}} \\
    &= \alpha\{ 1 + G(F_{mid})\epsilon + \cdots\}\frac{-1}{-\frac{1}{\sqrt{1-2\rho\zeta+\zeta^2}} \frac{\alpha}{\sigma_0}K^{-\beta}} \\
    &= \sigma_0K^{\beta}\{ 1 + G(F_{mid})\epsilon + \cdots\} \simeq \sigma_0K^{\beta} \quad \zeta_{F_0=K} = 0 \\
\end{align*}
$$

### LogNormal Model
From leture note#5 equation 51 we get the implied normal volatility and other expressions as:
$$
\begin{align*}
    \sigma_{ln}(T, K, F_0, \sigma_0, \alpha, \beta, \rho) &= \alpha\frac{log(F_0/K)}{D(\zeta)} \{ 1 + G(F_{mid})\epsilon + \cdots\} 
\end{align*}
$$

Follow the same calculation, we get:
$$
\begin{align*}
    \sigma_{ln}(T, K, F_0, \sigma_0, \alpha, \beta, \rho)_{ATM} &= \alpha\{ 1 + G(F_{mid})\epsilon + \cdots\} \frac{\frac{d(log(F_0/K))}{dK}}{\frac{dD(\zeta)}{dK}} \\

    &= \sigma_0K^{\beta-1}\{ 1 + G(F_{mid})\epsilon + \cdots\} \simeq \sigma_0K^{\beta-1}\\
\end{align*}
$$

```python
from abc import ABCMeta, abstractmethod
from enum import Enum
from math import log, sqrt
from scipy.optimize import newton, fmin_bfgs, minimize
```

```python
# Definition of Generic Option class with abstract method, value, impliedVol, impliedVolSABR
class OptionType(Enum):
    CALL = 1
    PUT = 2
    
class Option:
    def __init__(self, T, K, F_0, option_type):
        self.T = T
        self.K = K
        self.F0 = F_0
        self.option_type = option_type
        super().__init__()
    
    @abstractmethod
    def value(self, vol):
        pass
    
    @abstractmethod
    def impliedVol(self, price, init, tol):
        pass
    
    @abstractmethod
    def impliedVolSABR(self, alpha, beta, rho):
        pass
    
    def __str__(self):
        t = ""
        if(self.option_type is OptionType.CALL):
            t = "Call"
        else: 
            t = "Put"
        return t +' Option.\nMaturity: %s\nStrike: %s\nF0: %s\n' % (self.T, self.K, self.F0)
```

```python
# implementation of normal model
class normalmodel(Option):
    
    def __str__(self):
        return Option.__str__(self)
    
    @abstractmethod
    def value(self, vol):
        sigma_root_t = vol * sqrt(self.T)
        d = (self.F0 - self.K) / sigma_root_t
        d_minus = -d
        v = 0.0
        if(self.option_type is OptionType.CALL):
            v = sigma_root_t * (d * norm.cdf(d) + norm.pdf(d))
        else:
            v = sigma_root_t *(d_minus * norm.cdf(d_minus) + norm.pdf(d_minus))
        return v
    
    @abstractmethod
    def impliedVol(self, price, init, tol):
        implvol = newton(lambda x: (self.value(x) - price), init, tol = tol)
        return implvol
    
    @abstractmethod
    def impliedVolSABR(self, sigma, alpha, beta, rho):
        implvol = 0.0
        if self.K == self.F0:
            implvol = sigma * self.K **beta
        else:
            # zeta = alpha * (self.F0 - self.K)/sigma
            zeta = alpha * (self.F0**(1-beta) - self.K**(1-beta))/(sigma*(1-beta))
            delta = log((sqrt(1-2*rho*zeta+zeta**2)+zeta-rho)/(1-rho))
            # C(F) = F**beta
            # C'(F) = beta F**(beta-1)
            # C"(F) = beta(beta-1) F**(beta-2)
            F_mid = (self.F0 + self.K)/2
            gamma_1 = beta / F_mid
            gamma_2 = beta * (beta-1) / (F_mid**2)
            C_F_mid = F_mid**beta
            epsilon = self.T * alpha**2
            implvol = (alpha * (self.F0 - self.K)/delta) * (1 + epsilon*((2 * gamma_2 - gamma_1**2)*(sigma*C_F_mid/alpha)**2/24 + rho * gamma_1 * sigma * C_F_mid/(4*alpha) + (2-3*rho**2)/24) )
        return implvol
```

```python
# implementation of lognormal model
class lognormalmodel(Option):
    
    def __str__(self):
        return Option.__str__(self)
    
    @abstractmethod
    def value(self, vol):
        sigma_root_t = vol * sqrt(self.T)
        d = log(self.F0 / self.K) / sigma_root_t + 0.5 * sigma_root_t
        d_minus = log(self.F0 / self.K) / sigma_root_t - 0.5 * sigma_root_t
        v = 0.0
        if(self.option_type is OptionType.CALL):
            v = self.F0 * norm.cdf(d) - self.K * norm.cdf(d_minus)
        else:
            v = -self.F0 * norm.cdf(-d) + self.K * norm.cdf(-d_minus)
        return v
    
    @abstractmethod
    def impliedVol(self, price, init, tol):
        implvol = newton(lambda x: (self.value(x) - price), init, tol = tol)
        return implvol
    
    @abstractmethod
    def impliedVolSABR(self, sigma, alpha, beta, rho):
        implvol = 0.0
        if self.K == self.F0:
            implvol = sigma * self.K **(beta - 1)
        else:
            # zeta = alpha * (self.F0 - self.K)/sigma
            zeta = alpha * (self.F0**(1-beta) - self.K**(1-beta))/(sigma*(1-beta))
            delta = log((sqrt(zeta**2-2*rho*zeta+1)+zeta-rho)/(1-rho))
            # C(F) = F**beta
            # C'(F) = beta F**(beta-1)
            # C"(F) = beta(beta-1) F**(beta-2)
            F_mid = (self.F0 + self.K)/2
            gamma_1 = beta / F_mid
            gamma_2 = beta * (beta-1) / (F_mid**2)
            C_F_mid = F_mid**beta
            epsilon = self.T * alpha**2
            implvol = (alpha * log(self.F0/self.K)/delta) * (1 + epsilon*((2 * gamma_2 - gamma_1**2 + 1 / (F_mid**2))/24*((sigma*C_F_mid/alpha)**2) + rho * gamma_1 * sigma * C_F_mid/(4*alpha) + (2-3*rho**2)/24) )
        return implvol 
```

### (i) Generate Correlated Gaussian Random Numbers

 Explain the algorithm:

* First we assume two uncorrelated Gaussian random variables $Z_1, Z_2 \sim N(0,1)$. 

* We will have two new standard normal random variables $X$ and $Y$ with correlation $\rho$ by letting that

$$ X = Z_1$$

$$ Y = \rho Z_1+\sqrt{1-\rho^2}Z_2$$


* Obviously $X$ is still standard normal random variable. $Y$, a linear combination of two Guassians, is also Gaussian with $E[Y]$ = 0 and $Var[Y] = \rho^2 + 1 - \rho^2 = 1$. Thus $Y$ is also standard normal by definition.


* we need to verify the correlation is $\rho$.

$$ Corr(X,Y) = Corr(Z_1,\rho Z_1+\sqrt{1-\rho^2}Z_2) = \rho Corr(Z_1,Z_1)+0 = \rho $$
Thus, $X$ and $Y$ are two standard normal random variables with correlation $\rho$.

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# generate two dimensional Gaussian random numbers with correlation rho
def generate_correlated(numStep, numPath, rho):
    
    # numStep is the length of sequence
    # numPath is the number of paths
    # rho is the correlation
    z = np.random.normal(0,1,2*numStep*numPath)
    x = []
    y = []
    
    for i in range(numStep*numPath):
        x.append(z[2*i])
        y.append(rho*z[2*i]+np.sqrt(1-rho**2)*z[2*i+1])
    return x,y
```

### (ii) Write code for generating MC paths for the SABR model

```python
def MC_price(T,F0,K,sigma0,alpha,beta,rho,is_call, numStep, numPath):
    # T: maturity
    # F0: current value of the forward
    # sigma0: initial guess of market vol
    # alpha, beta, rho: model parameters
    # numStep: number of steps in discretization
    # numPath: number of paths
    # is_call: bool value for option, True for call and False for put
    # return: =price of the option
    
    x,y = generate_correlated(numStep, numPath, rho)
    dt = float(T)/float(numStep) #time step
    option_values = []
    
    for k in range(numPath):
        F = np.zeros(numStep+1)
        sigma = np.zeros(numStep+1)
        F[0] = F0
        sigma[0] = sigma0
        for i in range(numStep):  #discretization
            F[i+1] = max(F[i]+sigma[i]*np.power(F[i],beta)*np.sqrt(dt)*x[k*numStep+i],0)
            sigma[i+1] = sigma[i]*np.exp(alpha*np.sqrt(dt)*y[k*numStep+i]-alpha**2*dt/2)
        F_price=F[numStep]
        if is_call==True:
            option_values.append(max(F_price-K,0))
        else:
            option_values.append(max(K-F_price,0))
    return np.average(option_values)
```

```python
# Write the function for Black's lognormal model
def lognormal_price(T,F0,K,sigma0,alpha,beta,rho,is_call):
    implvol = 0.0
    if K == F0:
        implvol = sigma0 * K **(beta - 1)
    else:
        # zeta = alpha * (self.F0 - self.K)/sigma
        zeta = alpha * (F0**(1-beta) - K**(1-beta))/(sigma0*(1-beta))
        delta = np.log((np.sqrt(zeta**2-2*rho*zeta+1)+zeta-rho)/(1-rho))
        # C(F) = F**beta
        # C'(F) = beta F**(beta-1)
        # C"(F) = beta(beta-1) F**(beta-2)
        F_mid = (F0 + K)/2
        gamma_1 = beta / F_mid
        gamma_2 = beta * (beta-1) / (F_mid**2)
        C_F_mid = F_mid**beta
        epsilon = T * alpha**2
        implvol = (alpha * np.log(F0/K)/delta) * (1 + epsilon*((2 * gamma_2 - gamma_1**2 + 1 / (F_mid**2))/24*((sigma0*C_F_mid/alpha)**2) + rho * gamma_1 * sigma0 * C_F_mid/(4*alpha) + (2-3*rho**2)/24) )
    
    sigma_root_t = implvol * np.sqrt(T)
    d = np.log(F0 / K) / sigma_root_t + 0.5 * sigma_root_t
    d_minus = np.log(F0 / K) / sigma_root_t - 0.5 * sigma_root_t
    v = 0.0
    if(is_call == True):
        v = F0 * norm.cdf(d) - K * norm.cdf(d_minus)
    else:
        v = -F0 * norm.cdf(-d) + K * norm.cdf(-d_minus)
    return v
```

```python
# record the option value from MC simulation and black lognormal model
MC_lst = []
LogNormal_lst = []
numStep = 500
numPath = 5000

for idx in para.index:
    
    # set the parameters
    T = para.loc[idx]['T']
    K = para.loc[idx]['K']
    F0 = para.loc[idx]['F0']
    sigma0 = para.loc[idx]['sigma0']
    is_call = False
    if para.loc[idx]['Type'] == 'call':
        is_call = True
    alpha = para.loc[idx]['alpha']
    beta = para.loc[idx]['beta']
    rho = para.loc[idx]['rho']
    
    mc_price = MC_price(T, F0, K, sigma0, alpha, beta, rho, is_call, numStep, numPath)
    MC_lst.append(mc_price)
    Lognormal_price = lognormal_price(T, F0, K, sigma0, alpha, beta, rho, is_call)
    LogNormal_lst.append(Lognormal_price)

result = para
result['MC'] = MC_lst
result['LogNormal'] = LogNormal_lst
result['difference'] = result['MC']-result['LogNormal']
```
