* (i)

Plugging
$$\varphi(x)=\frac{1}{\sqrt{2\pi\sigma^2 T_{i-1}}x}\exp\left(-\frac{(\log(x/C_0)+\sigma^2T_{i-1}/2)^2}{2\sigma^2T_{i-1}}\right)$$
into $$V_{cpn}(0) = \mathcal{A}(0)\delta\int_0^\infty(ax^2+bx)\varphi(x)dx,$$
We get
$$V_{cpn}(0)=\mathcal{A}(0)\delta(aI+bII)$$
where
$$I = \int_0^\infty \frac{x}{\sqrt{2\pi\sigma^2 T_{i-1}}}\exp\left(-\frac{(\log(x/C_0)+\sigma^2T_{i-1}/2)^2}{2\sigma^2T_{i-1}}\right)dx$$
and
$$II = \int_0^\infty \frac{1}{\sqrt{2\pi\sigma^2 T_{i-1}}}\exp\left(-\frac{(\log(x/C_0)+\sigma^2T_{i-1}/2)^2}{2\sigma^2T_{i-1}}\right)dx$$
We first proceed to compute I. 

Let $y = \log(x/C_0)$ and rewrite I,
\begin{eqnarray}
I &=& C_0^2\int_{-\infty}^\infty\frac{1}{\sqrt{2\pi\sigma^2 T_{i-1}}}\exp\left(2y-\frac{(y+\sigma^2T_{i-1}/2)^2}{2\sigma^2T_{i-1}}\right)dy\\
&=& C_0^2 e^{\sigma^2T_{i-1}}\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi\sigma^2 T_{i-1}}}\exp\left(-\frac{(y-\frac{3}{2}\sigma^2T_{i-1})^2}{2\sigma^2T_{i-1}}\right)dy\\
&=& C_0^2 e^{\sigma^2T_{i-1}}
\end{eqnarray}
Likewise, 
$$II = C_0\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi\sigma^2 T_{i-1}}}\exp\left(-\frac{(y-\frac{1}{2}\sigma^2T_{i-1})^2}{2\sigma^2T_{i-1}}\right)dy=C_0$$
In summary, we conclude that
$$C_{cpn}(0)=\mathcal{A}(0)\delta(aC_0^2e^{\sigma^2T_{i-1}}+bC_0)$$

According to lecture notes #7 P26,

Integrate both sides of the above equation from $0$ to $k$, we get

Integrate it once again from $0$ to $K$,

$$V_{call}(K) = V_{call}(0)+\mathcal{A}(0)\int_0^K\int_0^k\varphi(x)dxdk+\mathcal{A}(0)\frac{\partial}{\partial K}V_{call}(0)K$$

Finally, we conclude that

$$

and

$$
\rho_S(X_1,X_2)=12\iint_{[0,1]\times[0,1]}C(u,v)dudv-3
$$

First let's calculate the double integral.

$$
\begin{aligned}
\iint_{[0,1]\times[0,1]}C(u_2,u_2)du_1du_2&=\int_{u_2=0}^{u_2=1}du_2\int_{u_1=u_2^{\frac{\alpha_2}{\alpha_1}}}^{u_1=1}u_1^{1-\alpha_1}u_2du_1+\int_{u_2=0}^{u_2=1}du_2\int_{u_1=0}^{u_1=u_2^{\frac{\alpha_2}{\alpha_1}}}u_1u_2^{1-\alpha_2}du_1\\
&=\frac1{2-\alpha_1}\int_0^1(u_2-u_2^{\frac{\alpha_2(2-\alpha_1)}{\alpha_1}+1})du_2+\frac12\int_0^1u_2^{1+\frac{2\alpha_2}{\alpha_1}-\alpha_2}du_2\\
&=\frac1{2-\alpha_1}[\frac12-\frac1{\frac{\alpha_2(2-\alpha_1)}{\alpha_1}+2}]+\frac12\cdot\frac1{2+\frac{2\alpha_2}{\alpha_1}-\alpha_2}\\
&=\frac{\alpha_1+\alpha_2}{2(2\alpha_1+2\alpha_2-\alpha_1\alpha_2)}
\end{aligned}
$$

Thus Spearman's $rho$ is

$$
\rho=12\iint_{[0,1]\times[0,1]}C(u_1,u_2)du_1du_2-3=\frac{3\alpha_1\alpha_2}{2\alpha_1+2\alpha_2-\alpha_1\alpha_2}
$$

### (iii)

First I think the lecture note has a typo on Lecture8, page 30, line 4.

$T_1,T_2,T_{12}$ are random variables, and thus they can no way follow Poisson processes. I think the professor actually means exponential distribution instead of Poisson process.

Similar to the one generating Gaussian copula, the algorithm to simulate $C_{MO}$ can be described as follows. Note that in this part, we use $\tilde{X}$ to denote a random variable, and $X$ to denote one of its realizations.

For each iteration:

Input: $\lambda_1,\lambda_2,\lambda_{12}$

1. generate $T_1\sim Exponential(\lambda_1)$, $T_2\sim Exponential(\lambda_2)$, $T_{12}\sim Exponential(\lambda_{12})$, and $(T_1,T_2,T_{12})$ are independent of each other.

2. Set $\tau_1=\min(T_1,T_{12}),\tau_2=\min(T_2,T_{12})$

3. Set

   $U_1=P(\tilde{\tau_1}>\tau_1)=e^{-(\lambda_1+\lambda_{12})\tau_1}$

   $U_2=P(\tilde{\tau_2}>\tau_2)=e^{-(\lambda_2+\lambda_{12})\tau_2}$
   
Output: $U_1,U_2$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def C_MO(lambd1, lambd2, lambd12, nSample):

    # step 1
    T1Vec = np.random.exponential(scale=1/lambd1, size=(nSample,1))
    T2Vec = np.random.exponential(scale=1/lambd2, size=(nSample,1))
    T12Vec = np.random.exponential(scale=1/lambd12, size=(nSample,1))

    data = pd.DataFrame(np.concatenate((T1Vec, T2Vec, T12Vec), axis=1),\
        index=np.arange(1, nSample+1), columns=['T1', 'T2', 'T12'])

    # step 2
    data['tau1'] = np.minimum(data['T1'], data['T12'])
    data['tau2'] = np.minimum(data['T2'], data['T12'])

    # step 3
    data['U1'] = np.exp(-(lambd1+lambd12)*data['tau1'])
    data['U2'] = np.exp(-(lambd2+lambd12)*data['tau2'])

    return data
```

```python
lambd1, lambd2, lambd12, nSample = 1.7, 0.7, 0.3, 2000

data = C_MO(lambd1, lambd2, lambd12, nSample)
data.head()
```

```python
plt.scatter(data['U1'], data['U2'], c=np.arctan2(data['U2'],data['U1']), alpha=0.5)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('U1')
plt.ylabel('U2')
plt.title('C_MO(u1, u2) with lambd1=1.7, lambda2=0.7,\nlambda12=0.3, and sample size=2000')
plt.show()
```

Let $x_i$ be the one day loss. By the definition of Var, $\Pr(x_i>VaR)=0.01 $. Let B be the number of total breaches within one year(250 days).

$$
\Pr(B=k)={250 \choose k} 0.01^k 0.99^{(250-k)}
$$

where $k=1,2,\dots,250$

The expectation of this binomial distribution is $np$.

Therefore

$$
E[B]=250\times 0.01=2.5
$$

(ii)

```python
from scipy.special import comb
def prob(k):
    return comb(250,k)*0.01**k*0.99**(250-k)
```

```python
for i in range(11):
    print("k = ",i,", probability is ",prob(i))
```
